```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
name = 'BoneCancer'
epic1 = pd.read_csv(name+'_degreecentrality.csv', delimiter=',')
epic3 = pd.read_csv(name+'_eigenvectorcentrality.csv', delimiter=',')
epic4 = pd.read_csv(name+'_clusteringCentrality.csv', delimiter=',')
#epic5 = pd.read_csv(name+'_enhancedInDegreeCentrality.csv', delimiter=',')
epic6 = pd.read_csv(name+'_closenesscentrality.csv', delimiter=',')
epic7 = pd.read_csv(name+'_khatzcentrality.csv', delimiter=',')
epic8 = pd.read_csv(name+'_betweennessscentrality.csv', delimiter=',')
plt.figure(figsize=(15, 5), dpi=80)
val1 = epic1['col2']
n1,x1,_1 = plt.hist(val1,bins=np.linspace(0, max(val1), 100), visible=False)
plt.plot(n1)
# val2 = epic2['col2']
# n2,x2,_2 = plt.hist(val2,bins=np.linspace(0.0001, max(val2), 100), visible=False)
# plt.plot(n2)
val3 = epic3['col2']
n3,x3,_3 = plt.hist(val3,bins=np.linspace(0, max(val3), 100), visible=False)
plt.plot(n3)
val4 = epic4['col1']
n4,x4,_4 = plt.hist(val4,bins=np.linspace(0, max(val4), 100), visible=False)
plt.plot(n4)
val5 = epic7['col2']
n5,x5,_5 = plt.hist(val5,bins=np.linspace(0, max(val5), 100), visible=False)
plt.plot(n5)
val6 = epic6['col2']
n6,x6,_6 = plt.hist(val6,bins=np.linspace(0, max(val6), 100), visible=False)
plt.plot(n6)
val8 = epic8['col2']
n8,x8,_8 = plt.hist(val8,bins=np.linspace(0, max(val8), 100), visible=False)
plt.plot(n8)

plt.xlabel('Bin Number')
plt.ylabel('Frequency')
plt.title('Centrality Distribution', fontsize=20)
plt.legend(['Degree', 'Eigenvector', 'Clustering', 'Closeness', 'Khatz', 'Betweenness'])
plt.grid()
plt.show()
```
