```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def allCentrality(name):
    epic1 = pd.read_csv(name+'_degreecentrality.csv', delimiter=',')
    epic3 = pd.read_csv(name+'_eigenvectorcentrality.csv', delimiter=',')
    epic4 = pd.read_csv(name+'_clusteringCentrality.csv', delimiter=',')
    #epic5 = pd.read_csv(name+'_enhancedInDegreeCentrality.csv', delimiter=',')
    epic6 = pd.read_csv(name+'_closenesscentrality.csv', delimiter=',')
    epic7 = pd.read_csv(name+'_khatzcentrality.csv', delimiter=',')
    epic8 = pd.read_csv(name+'_betweennessscentrality.csv', delimiter=',')
    plt.figure(figsize=(15, 5), dpi=80)
    val1 = epic1['col2']
    #print(val1.mean())
    n1,x1,_1 = plt.hist(val1,bins=np.linspace(0.001, max(val1), 100), visible=False)
    plt.plot(n1, color='r')
    val3 = epic3['col2']
    print(val3.mean())
    n3,x3,_3 = plt.hist(val3,bins=np.linspace(0.001, max(val3), 100), visible=False)
    plt.plot(n3, color='g')
    val4 = epic4['col1']
    #print(val4.mean())
    n4,x4,_4 = plt.hist(val4,bins=np.linspace(0.001, max(val4), 100), visible=False)
    plt.plot(n4, color='b')
    val5 = epic7['col2']
    print(val5.mean())
    n5,x5,_5 = plt.hist(val5,bins=np.linspace(0.001, max(val5), 100), visible=False)
    plt.plot(n5, color='m')
    val6 = epic6['col2']
    print(val6.mean())
    n6,x6,_6 = plt.hist(val6,bins=np.linspace(0.001, max(val6), 100), visible=False)
    plt.plot(n6, color='brown')
    val8 = epic8['col2']
    print(val8.mean())
    n8,x8,_8 = plt.hist(val8,bins=np.linspace(0.001, max(val8), 100), visible=False)
    plt.plot(n8, color='orange')

    plt.xlabel('Bin Number')
    plt.ylabel('Frequency')
    plt.title(name + ' Centrality Distribution', fontsize=20)
    plt.legend(['Degree', 'Eigenvector', 'Clustering', 'Closeness', 'Katz', 'Betweenness'])
    plt.grid()
    plt.show()
```

```python
allCentrality('BoneCancer')
allCentrality('BoneNormal')
allCentrality('BreastCancer')
allCentrality('BreastNormal')
allCentrality('ColonCancer')
allCentrality('ColonNormal')
allCentrality('KidneyCancer')
allCentrality('KidneyNormal')
allCentrality('LiverCancer')
allCentrality('LiverNormal')
```
